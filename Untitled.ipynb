```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
number_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
number_rating_df.rename(columns={'Book-Rating':'number_ratings'},inplace=True)
number_rating_df
```

```python
average_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
average_rating_df.rename(columns={'Book-Rating':'average_rating'},inplace=True)
average_rating_df
```

```python
popularity_df = number_rating_df.merge(average_rating_df,on='Book-Title')
popularity_df
```

```python
popularity_df = popularity_df[popularity_df['number_ratings']>=150].sort_values('average_rating',ascending=False).head(30)
```

```python
popularity_df = popularity_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','number_ratings','average_rating']]
```

```python
popularity_df['Image-URL-M'][0]
```

```python
#colaborative Filtering
```

```python
z = ratings_with_name.groupby('User-ID').count()['Book-Rating'] >= 150
userss = z[z].index
```

```python
rating_filter = ratings_with_name[ratings_with_name['User-ID'].isin(userss)]
```

```python
a = rating_filter.groupby('Book-Title').count()['Book-Rating']>=30
popular_books = a[a].index
```

```python
final_ratings = rating_filter[rating_filter['Book-Title'].isin(popular_books)]
```

```python
final = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
final.fillna(0,inplace=True)
```

```python
final
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similar_scores = cosine_similarity(final)
similar_scores.shape

```

```python
def recommend(book_chose):
    data = []
    array = np.where(final.index==book_chose)
    print(array)
    if array[0] != None:
        similar_items = sorted(list(enumerate(similar_scores[array[0][0]])), key=lambda x: x[1], reverse=True)[1:10]
        for i in similar_items:
            item = []
            temp_df = books[books['Book-Title'] == final.index[i[0]]]
            item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
            item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
            item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
            data.append(item)
        return data

  
```

```python
recommend('Zoya')

```
